```python
import pandas as pd
import numpy as np
import re
import webbrowser

from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
from nltk.corpus import stopwords
from wand.image import Image as WImage
from IPython.display import Image, display

# New libs for connection
import socket
import os
import ftplib
```

#### Python Server Socket

```python
class PdfConverter:

    def __init__(self, file_path):
        self.file_path = file_path
        
    # → Function used to convert pdf to text to allow the post-processing.
    def convert_pdf_to_txt(self):
        rsrcmgr = PDFResourceManager()
        retstr = StringIO()
        codec = 'utf-8'  # 'utf16','utf-8'
        laparams = LAParams()
        device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)
        fp = open(self.file_path, 'rb')
        interpreter = PDFPageInterpreter(rsrcmgr, device)
    
        # → Proposing just to read the first page of the pdf, because this is normally where the abstract and the introduction are.
        maxpages = 0

        caching = True
        pagenos = set()
        for page in PDFPage.get_pages(fp, pagenos, maxpages=maxpages, caching=caching, check_extractable=True):
            interpreter.process_page(page)
        fp.close()
        device.close()
        str = retstr.getvalue()
        retstr.close()
        return str
    
    def post_processing(self,raw_text):
        # → First post-processing for the text : Deleting all the symbols. [^\w] will match anything that's not alphanumeric or underscore.
        text = re.sub(r'[^\w]', ' ', raw_text)

        # → Second post-processing for the text. Avoiding stop words using the nltk dictionary.
        generalDictionary = stopwords.words('english')
        wordsArray = text.split()
        generalResult = [word for word in wordsArray if word.lower() not in generalDictionary]
        text = ' '.join(generalResult)

        # → Third post-processing for the text. Avoiding numbers.
        notNumbers = re.findall(r'[a-zA-Z]\w+',text)
        text = ' '.join(notNumbers)

        # → Printing the number of words in the text after the whole post-processing.
        #print(len(notNumbers))
        #print(text)
        
        return text
    
    # → Calculating the repetition frequency of each word.
    def get_keywords(self,post_processed_text):
        counts = dict()
        words = post_processed_text.split()

        for word in words:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1

        # → dictionary that contains the repetition frecuency of a word and its value.
        d = counts

        # → Simple array and counter to select the desired number of keywords that we want.
        array = []
        i = 0

        for w in sorted(d, key=d.get, reverse=True):
            array.append(w)
            # → Uncomment the following to the see the complete dictionary and its values.
            #print(w, d[w])

        desiredKeywords = 10
        keywords = []
        #print ('Keywords:\n')
        for i in range(desiredKeywords):
            #print (str(i+1)+' '+array[i])
            keywords.append(array[i])
        
        string_keywords = ' '.join(keywords)
        return string_keywords
```

```python
def ftp_server(filename):
    HOSTNAME = "192.168.192.150"
    USERNAME = "ethics4u"
    PASSWORD = "password"
    PATH = '/home/ethics4u/Papers/'
        
    ftp_server = ftplib.FTP(HOSTNAME, USERNAME, PASSWORD)
    ftp_server.encoding = "utf-8"
    ftp_server.cwd(PATH)
    
    local_filename = os.path.join(r"C:\Users\diego\Documents\MEGA\Education\Institut Polytechnique de Paris\Courses\Global Laboratory for Industry-Driven\Project\tsp-global-laboratory\project\temporal-storage", filename)
    lf = open(local_filename, "wb")
    ftp_server.retrbinary("RETR " + filename, lf.write, 8*1024)
    lf.close()
    ftp_server.quit()

def server_program():
    # → get the hostname
    host = '192.168.192.123'
    port = 5000  # initiate port above 1024

    server_socket = socket.socket()  # get instance
    server_socket.bind((host, port))  # bind host address and port together
    
    return server_socket
    
def connection_listen(server_socket):
    # → configure how many client the server can listen simultaneously
    server_socket.listen(2)
    conn, address = server_socket.accept()  # accept new connection
    
    #print("Connection from: " + str(address))
    while True:
        # → receive data stream. it won't accept data packet greater than 1024 bytes
        data_get = conn.recv(1024).decode()
        
        if not data_get:
            # if data is not received break
            connection_listen(server_socket)
        
        filename = data_get
        print("Host "+str(address)+" is requesting the following article: "+filename)
                
        # → downloading article
        ftp_server(filename)
        
        # → Calling pdfConverter class
        local_filepath='temporal-storage/'+filename
        pdfConverter = PdfConverter(local_filepath)    
        raw_text = pdfConverter.convert_pdf_to_txt().lower()

        # → Printing cover of the requested pdf
        with WImage(filename=local_filepath+"[0]") as img: img.save(filename="cover.png")
        display(Image(filename='cover.png'))
        
        # → Calling post_processing
        post_processed_text = pdfConverter.post_processing(raw_text)
        
        # → Calling post_processing
        keywords = pdfConverter.get_keywords(post_processed_text)
        data_send = keywords
        print("The following keywords have been sent: "+keywords+"\n")
    
        ###print("from connected user: " + str(data))
        #data_send = input(' -> ')
        conn.send(data_send.encode())  # send data to the client
        #os.remove(local_filename)
        
    conn.close()  # close the connection

if __name__ == '__main__':
    #while True:
    # → Starting connection with the front-end
    server_socket = server_program()
    connection_listen(server_socket)
```
